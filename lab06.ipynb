```python
import tensorflow as tf
from keras.models import Sequential, Model
from keras.layers import Input, Dense
```

```python
base_model = tf.keras.applications.MobileNetV2(
    input_shape=(160, 160, 3), include_top=False, weights="imagenet"
)

# We want to lock the existing model weights
base_model.trainable = False
```

```python
# creating a new model that we will train (it uses the old pretrained model without modifying)

inputs = tf.keras.Input(shape=(160, 160, 3))
x = tf.keras.layers.Flatten()(inputs)
outputs = tf.keras.layers.Dropout(0.2)(x)
model = tf.keras.Model(inputs, outputs)
```
